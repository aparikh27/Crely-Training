```python
import wfdb
import matplotlib.pyplot as plt

# 1. Download First 1500 Samples of Patient 100 from the MIT-BIH Arrhythmia Database
record = wfdb.rdrecord('100', pn_dir='mitdb', sampto=1500)
annotation = wfdb.rdann('100', 'atr', pn_dir='mitdb', sampto=1500)

# 2. Extract the signal and the labels
signal = record.p_signal[:, 0]  # Lead II 
peaks = annotation.sample       # Index of each heartbeat
labels = annotation.symbol      # Medical Label

# 3. Create the Plot
plt.figure(figsize=(12, 4))
plt.plot(signal, color='black', label='ECG Signal')
plt.scatter(peaks, [signal[p] for p in peaks], color='red', marker='x', label='Doctor Labels')

plt.title('Patient 100: Raw ECG with Annotations')
plt.xlabel('Samples')
plt.ylabel('Voltage (mV)')
plt.legend()
plt.show()
```
